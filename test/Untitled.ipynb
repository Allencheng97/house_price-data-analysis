```python
def cap(x,quantile=[0.01,0.99]):
    head,tail=x.quantile(quantile).values.tolist()
    return head,tail
```

```python
head,tail=cap(df['unit_price'])
```

```python
l=df[df['unit_price']<head].index.tolist()
```

```python
l+=df[df['unit_price']>tail].index.tolist()
```

```python
l
```

```python
df=df.drop(l)
```
